## Adding missing columns

```python
# Variables
e = 2.71828
kappa = 0.4
z0 = 0.008
```

```python
import numpy as np
```

```python
wind_profile['u_star10'] = (wind_profile['u10 (m/s)'] * kappa) / np.log(10/z0)
wind_profile['energy_10m'] = wind_profile['u10 (m/s)'] ** 3
wind_profile['energy_160m'] = wind_profile['u160 (m/s)'] ** 3
wind_profile.head()
```

```python
wind_profile.describe
```

## Plotting data

```python
wind_profile.head()
```

```python
import matplotlib.pyplot as plt

fig, (ax0, ax1) = plt.subplots(figsize=(12, 12),
                      nrows=2,
                      ncols=1,
                      sharex=False)

# Main figure title
fig.suptitle('Wind Profile Analysis', fontsize=16, fontweight='bold');


# Adding data to ax0
plot_0 = ax0.plot(wind_profile['time decimal hours'],
                     wind_profile['energy_10m'],
                     label='10m')
plot_0 = ax0.plot(wind_profile['time decimal hours'],
                 wind_profile['energy_160m'],
                 label='160m')

ax0.set(title='Energy Generation at 10m & 160m',
       xlabel='Time [h]',
       ylabel='Energy [kW]')

ax0.legend(loc='upper left');

ax0.axhline(y=wind_profile['energy_160m'].mean(),
              linestyle='--',
           color='red')

ax0.axhline(y=wind_profile['energy_10m'].mean(),
           linestyle='--',
           color='red');

# Adding data to ax1
plot_1 = ax1.scatter(x=wind_profile['time decimal hours'],
                 y=wind_profile['dir100'],
                 label='dir100')
plot_1 = ax1.scatter(x=wind_profile['time decimal hours'],
                 y=wind_profile['dir10'],
                 label='dir10')
ax1.set(title='Wind Direction at 10m & 160m',
       xlabel='Time [h]',
       ylabel='Direction')

ax1.legend(loc='upper left');
```
